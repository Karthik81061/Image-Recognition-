 # pre-processing of the image

```python
# tenserflow image
filename ="Image_Data//dog2.jpg"
```

```python

from tensorflow.keras.preprocessing import image
img = image.load_img(filename,target_size=(224,324))  ## load an Image
```

```python
plt.imshow(img)
```

```python
resized_img = image.img_to_array(img)
final_image = np.expand_dims(resized_img,axis=0) ## need fourth dim
final_image = tf.keras.applications.mobilenet.preprocess_input(final_image)
```

```python
final_image.shape
```

```python
pred = mobile2.predict(final_image)
```

```python
from tensorflow.keras.applications import imagenet_utils
```

```python
results = imagenet_utils.decode_predictions(pred)
```

```python
print(results)
```

```python
final_image1 = tf.keras.applications.mobilenet_v3.preprocess_input(final_image)
```

```python
pred2 = mobile2.predict(final_image1)
```
